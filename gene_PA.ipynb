```python
import pandas as pd
from clustergrammer2 import *
import numpy as np

dataframe = pd.read_parquet('gene_presence_absence.parquet')
net = Network(CGM2)

rows = input('choose number of rows ')
cols = input('choose number of columns ')
if rows is not None:
    if cols is not None:
        www = dataframe.iloc[:int(rows),0:int(cols)]
        try:
            www += 1
        except:
            www.set_index('Unnamed: 0', inplace=True)
            www += 1
            
        www += 1

        net.load_df(www)
        net.cluster(enrichrgram=True)
net.widget()
```
